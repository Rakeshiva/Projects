```python
import numpy as np
import pandas as pd
url='https://raw.githubusercontent.com/SR1608/Datasets/main/covid-data.csv'
df=pd.read_csv(url)

##High level understanding
df.shape       #shape of the dataset/(no.of rows and no of columns info)

df.dtypes  #Data types of columns

df.info()  
df.describe()  #Info & describe of data in dataframe

df.columns

df['location'].nunique() #Gives the count of unique values in location column

df['continent'].mode() #Continent which has maximum frequency using values 


df['continent'].value_counts()

print("Maximum no.of 'total_cases----",df['total_cases'].max())
print("Average of 'total_cases-------",df['total_cases'].mean())


df.total_deaths.quantile([0.25,0.50,0.75])

df.groupby('continent').agg({'human_development_index':'max'})

df.groupby('continent').agg({'gdp_per_capita':'min'})

df.groupby('continent').agg({'total_deaths':'max'})

fd=df[['continent','location','date','total_cases','total_deaths','gdp_per_capita','human_development_index']]

fd #fd=filtered data

fd.duplicated().sum()

fd.info()

a=fd.dropna(subset=['continent']) # Remove all observations where continent column value is missing
data=pd.DataFrame(a)                        

data.isna().sum()

data['total_deaths']=data['total_deaths'].fillna(0)

data['total_cases']=data['total_cases'].fillna(0)

data['gdp_per_capita']=data['gdp_per_capita'].fillna(0)

data['human_development_index']=data['human_development_index'].fillna(0)

data.isna().sum()

data['date']=pd.to_datetime(data['date'])

data.head()

data['month']=data['date'].dt.month

data.head()

data.info()

data.convert_dtypes()

data['total_cases']=data['total_cases'].astype('float64')

data.info()

df_groupby=data.groupby(['continent']).max()

df_groupby.reset_index()

df_groupby['total_deaths_to_total_cases']=df_groupby['total_deaths']/df_groupby['total_cases']

df_groupby.reset_index()

df_groupby.info()

import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns

####Univariate analysis on 'gdp_per_capita' 

sns.distplot(df_groupby['gdp_per_capita'],kde=False)

##Plot a scatter plot of 'total_cases' & 'gdp_per_capita'

sns.jointplot(data=df_groupby, x='total_cases' , y ='gdp_per_capita',kind='scatter',hue='continent')

sns.pairplot(data=df_groupby)

df_groupby.reset_index()

df_groupby.info()

df_groupby.reset_index(inplace=True)

plt.rcParams['figure.figsize']=(30,20)
sns.catplot(data=df_groupby , x='continent' , y='total_cases' , kind ='bar')

from matplotlib.pyplot import figure
figure( figsize=(10, 8))

sns.catplot(data=df_groupby , x='continent' , y='total_cases' , kind ='bar',height=5,aspect=2)

df_groupby.to_csv

my_df.set_index('Person')
```
